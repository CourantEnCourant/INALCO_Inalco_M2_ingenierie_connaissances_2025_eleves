# TP3 Le retour de l'agent

image.png

Les agents IA sont le "sujets chaud" de l'industrie en 2025.
Savoir comment un agent IA fonctionne, c'est bien, mais savoir en implémenter un soi même, c'est mieux et pourrait vous donner un coup de boost pour vos entretiens.


Il ne s'agit pas de "réinventer la roue" et vous faire coder les mécaniques des agents IA mais de voir comment vous pouvez les utiliser pour créer un projet pratique utilisant un ou des agents IA.
Pour cela je vous propose de créer un projet à base d'agents IA en utilisant la librairie smolagent de Hugging Face.
Ce choix est justifié par plusieurs raisons, la première est que ce framework est français. De plus l'emploi de la librairie permet un nombre quotidien d'appel gratuit à un bon LLM (Qwen 2.5 32B) ce qui vous permettra de développer gratuitement et dans de bonnes conditions. Enfin ce framework est minimaliste et permet rapidement de coder des choses fonctionnelles sans avoir à recourir à une trop grande abstraction.


Votre tâche : utiliser des agents pour automatiser une tâche en lien avec l'ingénierie des connaissances ou d'autres notions vues en cours (graphes, domaine de spécialité, l'article sur lequel vous avez travaillé ...).  Votre système devra être accessible depuis une interface graphique ergonomique et claire.
